```python
# –––––––––––– 1º Step: ––––––––––––

#Take a list of compound names and group them by the last word.
# This way we'll have more compact dataset, easier to scan later.

# For example, we have:
#   performer
#   Burt Reynolds
#   Debbie Reynolds
#   Ryan Reynolds

# And turn it into:
#   Reynolds:	Debbie, Ryan, Burt


#Import modules 
import sys
import re
import csv
csv.field_size_limit(sys.maxsize)

# Create the dictionary that will store the results with the names grouped by last word
  
results={}

# Open the initial TSV file that stores a simple list 

with open("./data/ready-to-use/a-simple-list-of-performers.tsv", "r", encoding="utf8") as file:
    simple_list = csv.DictReader(file, delimiter="\t", quoting=csv.QUOTE_NONE)

    # For every item in the list
    for line in simple_list:                        

        #Make a text string with every result:
        performer=str(line.get('performer'))

        # If there is no blank space in that text string, it means it is a single result.
        # So you cannot blend it.
        # You cannot blend Beyoncé, okay?
        if ' ' not in performer: 
            pass 
        
        #But, if there is a blank space, we can procceed
        else: 
            
            #Create two variables from every item: 
            name, last_name = performer.rsplit(' ', 1)

            # If the last name has not been included yet in the dictionary with the grouped list  
            if last_name not in results:
                
                # Add it and include the name we are checking right now 
                results[last_name]=name

            # But if the last name already exists in that dictionary…
            else:

                # Make a variable with a text string from that last name
                last_name_to_check=str(results[last_name])

                #If the first name is already included in the group
                if re.search(rf'\b{name}\b', last_name_to_check):

                    #Do nothing. So you avoid checking duplicate items from the initial list
                    pass

                #Otherwise…
                else:
                
                    # Add to that group (dictionary key) the new name 
                    results.update({last_name:(results[last_name])+ ", " +name})


# Now we store all the grouped results in a TSV file. 
with open("./data/ready-to-use/a-grouped-list-of-performers.tsv", "w", encoding="utf8") as file:
    file.write('last_name' + '\t' + 'name' + '\n')
    iterable=0
    for i in results:
        key = list(results.keys())[iterable]
        file.write(key + '\t' + str(results[i]) + '\n')
        iterable+=1
```

```python
# –––––––––––– 2º Step: ––––––––––––

# Take the TSV file with the grouped list and check every value in every group.
# Check if there are suitable results to be blended into new names.
# Store every new name in a tsv file.


# Pick a group.
# Check every value in that group.
# If there is a second group named as that value: 
#   Pick every value from that second group.
#   Make a new name by adding that value, the name of its group and the name of the first group.

"""
For example:

We are checking the value 'Ryan' from the group 'Reynolds' 
  Reynolds: Debbie, Burt, Ryan

And we find this group: 
  Ryan: Amy, Meg

Now we make:
  Amy Ryan Reynolds
  Meg Ryan Reynolds
"""

#Import modules
import sys
import csv

csv.field_size_limit(sys.maxsize)

# Create a list with the new names 
new_names = []

# Open the initial TSV file that stores a list of grouped items 
with open('./data/ready-to-use/a-grouped-list-of-performers.tsv', 'r') as file:
  reader = csv.reader(file, delimiter='\t', quoting=csv.QUOTE_NONE)
  names_dict = {}

  # Look into every row of the csv object:
  for row in reader:

    # If the first row of the list is a header.
    if row[0] == 'lastName':

      #Do nothing. There is nothing to check in there.
      continue  


    names = [name.strip() for name in row[1].split(',')]

    names_dict[row[0]] = names


for last_name, names in names_dict.items():

  for name in names:

    if names_dict.get(name) is not None:

      for name3 in names_dict[name]:
        new_name= f"{name3} {name} {last_name}"
        new_names.append(new_name)  
    

with open("./data/results/a-list-of-blended-performers.tsv", "w", encoding="utf8") as file:
  file.write('performer' + '\t' + '\n')
  iterable=0
  for name in new_names:
    file.write(new_names[iterable] + '\t' + '\n')
    iterable+=1

                
```
